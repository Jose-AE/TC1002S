### 1.Carga de datos con pandas

```python
import pandas as pd

df = pd.read_csv('/content/avocado.csv')
```

```python
#Verificamos contenido del dataset
df.shape
```

```python
#Revisamos las columnas y hacemos
df.columns
```

```python
#Revisamos que valores contienen las columnas
df.dtypes
```

```python
#Aseguramos que las columnas no contegan valores nulos con df.dropna
df = df[
    [
        'Date', 'AveragePrice', 'Total Volume', '4046', '4225',
       '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags', 'type',
       'year', 'region']
    ]
df.dropna()
```

```python
#Analizar las variables para saber que representan y sus rangos
print(df.describe())

print(f"La mediana de los datos son: \n{df.median()}")
```
